## 데이터 추출(발급회원번호 기준으로)
- 데이터 로드(vip , general 따로)
- 발급회원번호 txt 파일 로드
- 발급회원번호 기준으로 데이터 추출
- 새로운 데이터프레임으로 변수 저장

```python
import pandas as pd
import os

# 1. 데이터 로드 (vip, general 따로)
vip_path = '../../data/통합데이터/vip_customers_all.csv'  # 경로 수정 필요
general_path = '../../data/통합데이터/general_customers_all.csv' # 경로 수정 필요

# 파일 존재 여부 확인 후 로드
if os.path.exists(vip_path):
    df_vip = pd.read_csv(vip_path)
    print(f"VIP 데이터 로드 완료: {df_vip.shape}")
else:
    print(f"VIP 파일 없음: {vip_path}")

if os.path.exists(general_path):
    df_general = pd.read_csv(general_path)
    print(f"General 데이터 로드 완료: {df_general.shape}")
else:
    print(f"General 파일 없음: {general_path}")
```

```python
# 2. 발급회원번호 txt 파일 로드
# TODO: 발급회원번호 리스트 파일이 준비되면 아래 경로를 수정하고 주석을 해제하세요.
vip_id_file_path = '../251217/1차_VIP_2000.txt'
general_id_file_path = '../251217/1차_general_2000.txt'

vip_ids = []
general_ids = []

# VIP ID 로드 (주석 처리됨)
if os.path.exists(vip_id_file_path):
    with open(vip_id_file_path, 'r') as f:
        vip_ids = [line.strip() for line in f.readlines()]
    print(f"VIP 발급회원번호 리스트 로드 완료: {len(vip_ids)}명")
else:
    print("VIP 발급회원번호 파일을 찾을 수 없습니다.")

# General ID 로드 (주석 처리됨)
if os.path.exists(general_id_file_path):
    with open(general_id_file_path, 'r') as f:
        general_ids = [line.strip() for line in f.readlines()]
    print(f"General 발급회원번호 리스트 로드 완료: {len(general_ids)}명")
else:
    print("General 발급회원번호 파일을 찾을 수 없습니다.")
```

```python
# 3. 발급회원번호 기준으로 데이터 추출

# VIP 데이터 필터링
if vip_ids:
    if 'df_vip' in locals():
        df_vip_filtered = df_vip[df_vip['발급회원번호'].isin(vip_ids)]
        print(f"VIP 데이터 필터링 완료: {df_vip_filtered.shape}")
else:
    print("필터링할 VIP ID가 없습니다. (ID 리스트 로드 필요)")
    # 테스트를 위해 원본 데이터를 그대로 필터링된 데이터 변수에 할당할 수도 있습니다.
    if 'df_vip' in locals(): df_vip_filtered = df_vip.copy()

# General 데이터 필터링
if general_ids:
    if 'df_general' in locals():
        df_general_filtered = df_general[df_general['발급회원번호'].isin(general_ids)]
        print(f"General 데이터 필터링 완료: {df_general_filtered.shape}")
else:
    print("필터링할 General ID가 없습니다. (ID 리스트 로드 필요)")
    # 테스트를 위해 원본 데이터를 그대로 필터링된 데이터 변수에 할당할 수도 있습니다.
    if 'df_general' in locals(): df_general_filtered = df_general.copy()
```

## 데이터 확인
- 데이터프레임 확인
- 데이터프레임의 컬럼 확인
- 데이터프레임의 특정 컬럼의 데이터 확인

```python
# 4. 데이터프레임 확인

print("=== VIP 데이터프레임 확인 ===")
if 'df_vip_filtered' in locals():
    display(df_vip_filtered.head())
    print("\n[VIP 컬럼 정보]")
    print(df_vip_filtered.info())
    print("\n[VIP 컬럼 목록]")
    print(df_vip_filtered.columns.tolist())

print("\n=== General 데이터프레임 확인 ===")
if 'df_general_filtered' in locals():
    display(df_general_filtered.head())
    print("\n[General 컬럼 정보]")
    print(df_general_filtered.info())
    print("\n[General 컬럼 목록]")
    print(df_general_filtered.columns.tolist())
```

```python
# vip, general 데이터프레임 csv 파일로 따로 저장
df_vip.to_csv('1차_vip_2000.csv', index=False)
df_general.to_csv('1차_general_2000.csv', index=False)
```
