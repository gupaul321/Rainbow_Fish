```python
import pandas as pd 
import numpy as np
```

```python
df_vif = pd.read_csv("../12.17/1cha_VIP(shap_vif)_74.csv")
```

```python
df_vif.head()
```

```python
# 2. 'vif' 컬럼에서 무한대(inf) 값 확인
# 판다스는 문자열 'inf'나 실제 np.inf를 모두 처리할 수 있어야 해.
print("필터링 전 전체 변수 개수:", len(df_vif))
inf_count = df_vif[df_vif['vif'] == np.inf].shape[0]
print(f"VIF가 inf인 변수 개수: {inf_count}개")
```

```python
# 3. VIF 지수가 inf(무한대)인 값 제외하기
# np.isinf 함수를 사용하면 숫자형 inf를 깔끔하게 골라낼 수 있어.
df_vif_filtered = df_vif[~df_vif['vif'].isin([np.inf, float('inf')])]

# 만약 데이터에 9999 같은 특정 큰 숫자로 기록되어 있다면 아래처럼 추가로 지울 수도 있어.
df_vif_filtered = df_vif_filtered[df_vif_filtered['vif'] < 9999]

# 4. 결과 확인 및 CSV 파일로 저장
print("필터링 후 남은 변수 개수:", len(df_vif_filtered))
df_vif_filtered.to_csv('1cha_VIP(shap_vif)_vifFiltered_55.csv', index=False, encoding='utf-8-sig')
```
