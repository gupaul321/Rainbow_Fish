```python
from functools import reduce

dfs = [df1, df2, df3, df4, df5, df6, df7]

# 2. 반복문을 사용해서 순서대로 합치기
# 'how=left'는 df1(회원정보)을 기준으로 데이터를 붙인다는 뜻이야.
# 만약 모든 데이터를 다 살리고 싶으면 how='outer'로 바꾸면 돼!
df_final = dfs[0]

for df in dfs[1:]:
    df_final = pd.merge(df_final, df, on=['발급회원번호', '기준년월'], how='left')

# 3. 잘 합쳐졌는지 확인해볼까?
print("최종 데이터 크기:", df_final.shape)
print("\n[상위 5개 행 미리보기]")
display(df_final.head())

# 4. 혹시 모르니 컬럼명 중복 확인 (Merge 과정에서 _x, _y가 생겼는지 체크)
print("\n[전체 컬럼 리스트]")
print(df_final.columns.tolist())
```
