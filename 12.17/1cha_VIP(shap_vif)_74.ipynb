```python
import pandas as pd

df = pd.read_csv('1차_VIP_shap_vif.csv')

# 2. 상위 74개 행만 자르기 (인덱스 0 ~ 73)
top74_df = df.head(74)
# 또는 df.iloc[:74] 도 같은 결과야!

# 3. 새로운 CSV 파일로 저장
filename = '1cha_VIP(shap_vif)_74.csv'
top74_df.to_csv(filename, index=False)

print(f"[완료] '{filename}' 파일이 생성되었습니다!")
print(f"추출된 변수 개수: {len(top74_df)}개")

# 확인 차 상위 5개와 마지막 5개(70~74위) 출력
print("\n[상위 5개]")
display(top74_df.head())
print("\n[하위 5개 (70~74위)]")
display(top74_df.tail())
```
