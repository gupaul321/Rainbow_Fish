```python
import pandas as pd

# 1. 데이터 불러오기
# df = pd.read_csv('../../30만원본/카드성과정보_300k.csv', encoding='cp949')

# 2. 월 정보 추출 (예: 201807 -> '07')
# 컬럼명 뒤에 붙일 접미사(Suffix)를 만듭니다.
df['month_suffix'] = df['기준년월'].astype(str).str[-2:]

# 3. 피벗(Pivot) 수행
# index: 고정할 기준 (발급회원번호)
# columns: 컬럼으로 보낼 값 (month_suffix -> 07, 08...)
# values: 나머지 모든 변수들
feature_cols = [c for c in df.columns if c not in ['발급회원번호', '기준년월', 'month_suffix']]

# pivot 함수를 사용하여 데이터 형태 변환
df_pivoted = df.pivot(index='발급회원번호', columns='month_suffix', values=feature_cols)

# 4. 컬럼 이름 평탄화 (MultiIndex -> Single Index)
# 예: ('이용금액', '07') -> '이용금액_07' 형태로 변경
df_pivoted.columns = [f'{col}_{month}' for col, month in df_pivoted.columns]

# 5. 인덱스 초기화 (발급회원번호를 다시 컬럼으로)
df_pivoted = df_pivoted.reset_index()

# 6. 결과 확인 및 저장
print(f"변환 전 데이터 크기: {df.shape}")
print(f"변환 후 데이터 크기: {df_pivoted.shape}")
# df_pivoted.to_csv('카드회원정보_WideFormat.csv', index=False)
```
