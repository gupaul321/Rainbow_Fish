```python
import re

def expand_features(feature_list):
    expanded = set()
    months = ['07', '08', '09', '10', '11', '12']
    
    for feat in feature_list:
        # 1. 컬럼명 끝이 _07, _08, ..., _12로 끝나는지 확인
        match = re.search(r'_(07|08|09|10|11|12)$', feat)
        
        if match:
            # 숫자를 뗀 앞부분(Prefix) 추출 (예: '이용건수_일시불_B0M')
            prefix = feat[:match.start()]
            # 해당 앞부분에 대해 07~12를 모두 생성
            for m in months:
                expanded.add(f"{prefix}_{m}")
        else:
            # 숫자가 없거나 형식이 다르면 그대로 유지
            expanded.add(feat)
            
    return sorted(list(expanded))

# 혜민님의 99개 리스트를 여기에 넣으세요!
final_feature_list = expand_features(selected_features) # selected_features는 아까 그 99개 변수

print(f"✅ 확장 전 컬럼 수: {len(selected_features)}개")
print(f"🚀 확장 후 컬럼 수: {len(final_feature_list)}개")

final_feature_list

```
